```python
a=1
while(a<=9):
  b=1
  while(b<=9):
    if (a*b)>=10:
      print(f'{b}x{a}={a*b}',end='  ')
    else:
      print(f'{b}x{a}={a*b}',end='   ')
    b=b+1
  print('')
  a=a+1
```

```python
import random
ans=random.sample(range(0,10),4)
print(ans)
num=0
a=0
while(a!=4):
  a=b=n=0
  num +=1
  guess=list(input('請輸入四個不重複之數字'))
  if len(guess)>4:
    print('輸入數字過多，請輸入四個不重複數字')
    print('')
    num +=1
  else:
    for i in guess:
      if int(guess[n]) == ans[n]:
        a += 1
      else:
        if int(i) in ans:
          b += 1
      n += 1
    output=','.join(guess).replace(',','')
    print(f'{output}:{a}A{b}B')
    print('')
print(f'猜對了！ 總共用了{num}次')
```
